```python
with open('parsed_dict', 'r') as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data.items(), columns=['entry', 'options'])
```

```python
df
# В options последним элементом временно хранится номер страницы
```

```python
pages = df['options'].apply(lambda x: x[-1])
df['page'] = pages
df['options'] = df['options'].apply(lambda x: sum([el.strip('.?').split('; ') for el in x[:-1]], []))
df.sort_values(by=['page'], inplace=True)
df.set_index('page', inplace=True)
```

```python
df
```

```python
df = df[df['options'].apply(lambda x: len(x) > 0)]
df
```

```python
df[df['options'].apply(lambda x: len(x) > 1)]
```

```python
dictionary = df.to_dict('index')
rows = []
for item in dictionary.items():
  if len(item[1]['options']) > 1:
    for i in range(1, len(item[1]['options'])):
        rows.append(f'{item[1]["entry"]}\t{item[1]["options"][0]}\t{item[1]["options"][i]}\tCymraeg\n')
```

```python
with open('Cymraeg_dict.txt', 'w', encoding='utf-8') as file:
  for row in rows:
    file.write(row)
```
